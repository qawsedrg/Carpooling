```python
month2 = pd.read_csv('yellow_tripdata_2019-02.csv')
month3 = pd.read_csv('yellow_tripdata_2019-03.csv')
month4 = pd.read_csv('yellow_tripdata_2019-04.csv')
month5 = pd.read_csv('yellow_tripdata_2019-05.csv')
month6 = pd.read_csv('yellow_tripdata_2019-06.csv')
month7 = pd.read_csv('yellow_tripdata_2019-07.csv')
month8 = pd.read_csv('yellow_tripdata_2019-08.csv')
month9 = pd.read_csv('yellow_tripdata_2019-09.csv')
month10 = pd.read_csv('yellow_tripdata_2019-10.csv')
month11 = pd.read_csv('yellow_tripdata_2019-11.csv')
month12 = pd.read_csv('yellow_tripdata_2019-12.csv')
```

```python
def process(DF,MM):
    DF['congestion_surcharge'] = DF['congestion_surcharge'].fillna(0)
    DF['tpep_pickup_datetime'] = pd.to_datetime(DF['tpep_pickup_datetime']) #%%change to date format
    DF.loc[:, 'tpep_pickup_date'] = DF['tpep_pickup_datetime'].dt.date
    DF.loc[:, 'tpep_pickup_hour'] = DF['tpep_pickup_datetime'].dt.hour
    DF['tpep_dropoff_datetime'] = pd.to_datetime(DF['tpep_dropoff_datetime'])
    DF.loc[:, 'tpep_dropoff_hour'] = DF['tpep_dropoff_datetime'].dt.hour
    DF.loc[:, 'tpep_pickup_year'] = DF['tpep_pickup_datetime'].dt.year
    DF.loc[:, 'tpep_pickup_month'] = DF['tpep_pickup_datetime'].dt.month
    DF['trip_duration'] = (DF['tpep_dropoff_datetime'] - DF['tpep_pickup_datetime']).dt.total_seconds()
    DF['average_speed'] = (DF['trip_distance']/DF['trip_duration']*3600)
    DF.drop(DF[(DF['total_amount']<=0)|(DF['total_amount']>300)].index, inplace=True)
    DF.drop(DF[DF['fare_amount']<=0].index,inplace=True)
    DF.drop(DF[(DF['trip_distance']>200)|(DF['trip_distance']<0.1)].index,inplace=True)
    DF = DF[(DF.tpep_pickup_month==MM)&(DF.tpep_pickup_year==2019)]
    DF.drop(DF[(DF['trip_duration']>10000)|(DF['trip_duration']<60)].index,inplace=True)
    DF.drop(DF[DF['average_speed']>100].index,inplace=True)
    m2_C = DF.groupby('tpep_pickup_date').count()['VendorID']
    m2_avg = DF.groupby('tpep_pickup_date').mean().average_speed
    m2_C.to_csv('m'+str(MM)+'_C.csv')
    m2_avg.to_csv('m'+str(MM)+'_avg.csv')
    
    return
```

```python
process(month2,2)
```

```python
for i,j in zip([month2,month3,month4,month5,month6,month7,month8,month9,month10,month11,month12],
               list(range(2,13))):
    process(i,j)
```
